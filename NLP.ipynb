```python
from nltk.stem import PorterStemmer, LancasterStemmer

st1 = PorterStemmer()
st2 = LancasterStemmer()

words = ["fly", "flies", "flying", "flew", "flown"]

print("PorsterStemmer    :", [st1.stem(w) for w in words])
print("LancasterStemmer  :", [st2.stem(w) for w in words])
```

```python
from nltk.stem import WordNetLemmatizer

lm = WordNetLemmatizer()

[lm.lemmatize(w, pos="v") for w in words]
```

```python
import nltk
from nltk import Text

# Sample text
text = "Hello, how are you doing today? I hope everything is going well."

# Tokenize the text
tokens = nltk.word_tokenize(text)

# Create an NLTK Text object
text_object = Text(tokens)

# Display some information about the text
print(text_object)

# Example: Concordance
text_object.concordance("you")
```
